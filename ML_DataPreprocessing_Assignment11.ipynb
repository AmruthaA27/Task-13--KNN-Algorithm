**Step 1: Importing the libraries**

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

**Step 2: Importing dataset**

```python
df_data = pd.read_csv('Data.csv')
df_data
```

**Step 3: Handling the missing data**

```python
df_data["Salary"] = df_data["Salary"].replace(np.NaN,df_data["Salary"].mean())
df_data["Salary"] = df_data["Salary"].astype(float).round(decimals=2)
df_data["Age"] = df_data["Age"].replace(np.NaN,df_data["Age"].mean())
df_data["Age"] = df_data["Age"].astype(int)
df_data
```

**Step 4: Encoding categorical data**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label = le.fit_transform(df_data['Country'])
df_data.drop("Country", axis=1, inplace=True)
df_data["Country"] = label
df_data
```

**Step 5: Creating a dummy variable**

```python
X = df_data[["Country","Age","Salary"]]
Y = df_data[['Purchased']]
X
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state = 4)
```

**Step 7: Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train
```
